```python
import pandas as pd
df = pd.read_csv('V2.csv', encoding='utf-8', delimiter=';')
del df[df.columns[0]]
df["Class"] = df["Class"] - 1
df.head()
```

```python
replace_map = {
    "Status" :
    {
        "<0 DM": 1,
        "0-200 DM": 2,
        ">=200 DM": 3,
        "no checking account": 4
    },

    "Credit.history":
    {
        "no credits taken/all credits paid back duly": 1,
        "all credits at this bank paid back duly": 2,
        "existing credits paid back duly till now": 3,
        "delay in paying off in the past": 4,
        "critical account/other credits existing (not at this bank)": 5
    },

    "Purpose":
    {
        "car (new)": 1,
        "car (used)": 2,
        "furniture/equipment": 3,
        "radio/television": 4,
        "domestic appliances": 5,
        "repairs": 6,
        "education": 7,
        "(vacation - does not exist?)": 8,
        "retraining": 9,
        "business": 10
    },

    "Savings.account":
    {
        "<100 DM": 1,
        "100-500 DM": 2,
        "500-1000 DM": 3,
        ">=1000 DM": 4,
        "unknown/ no savings account": 5
    },

    "Personal.status.and.sex":
    {
        "male: divorced/separated": 1,
        "female: divorced/separated/married": 2,
        "male: single": 3,
        "male: married/widowed": 4
    },

    "Job":
    {
        "unemployed/ unskilled  - non-resident": 1,
        "unskilled - resident": 2,
        "skilled employee / official": 3,
        "management/ self-employed/highly qualified employee/ officer": 4
    },

    "foreign.worker":
    {
        "yes": 1,
        "no": 2
    }
}
data_coded = df.copy()
data_coded.replace(replace_map, inplace=True)
```

```python
corr_matrix_in = data_coded.drop("Class", axis=1).corr(method="pearson")
corr_matrix_in
```

```python
cols = list(data_coded.columns)
cols.remove("Class")
corr_matrix_flag = data_coded.corr(method="pearson").drop(cols, axis=1).drop("Class", axis=0)
corr_matrix_flag
```

```python
corr_matrix_in.loc["Number.of.people"]["Installment.rate"]
```

```python
from itertools import chain, combinations
def powerset(iterable):
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1))
combinations = list(powerset(data_coded.drop("Class", axis=1).columns))
combinations = list(map(lambda x: list(x), combinations))
```

```python
def FSK(R, r):
    import numpy as np
    R = np.abs(R)
    r = np.abs(r)
    k = len(R)
    sum_R = R.sum()
    sum_r = r.sum()
    res = sum_R / (k + 2*sum_r)**(1/2)
    return res
```

```python
import numpy as np
FSK_s = []
new_combs = []
for comb in combinations:
    R = corr_matrix_flag.loc[comb].values.flatten()
    r = corr_matrix_in.loc[comb][comb].values.flatten()
    r = np.unique(r[r != 1])
    if len(r) is 0: continue
    FSK_s.append(FSK(R, r))
    new_combs.append(comb)
```

```python
index = np.argmax(FSK_s)
selected = new_combs[index]
```

```python
selected
```

# вторая часть

```python
train = data_coded[:900]
test = data_coded[900:]

X_train = train[selected]
X_test = test[selected]
Y_train = train["Class"].values
Y_test = test["Class"].values
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
reg_lin = LinearRegression().fit(X_train, Y_train)
y_pred = reg_lin.predict(X_test)
y_pred = list(map(lambda x: 1 if x >= 0.5 else 0, y_pred))
```

```python
TP = 0
TN = 0
FP = 0
FN = 0
for i in range(len(y_pred)):
    pred = y_pred[i]
    test = Y_test[i]
    if test == 1 and pred == 1:
        TP += 1
    elif test == 0 and pred == 0:
        TN += 1
    elif test == 0 and pred == 1:
        FP += 1
    elif test == 1 and pred == 0:
        FN += 1
print ("TP", TP)
print ("TN", TN)
print ("FP", FP)
print ("FN", FN)
```

```python
from sklearn.linear_model import LogisticRegression
reg_log = LogisticRegression().fit(X_train, Y_train)
y_pred = list(map(lambda x: x[1], reg_log.predict_proba(X_test)))
y_pred = list(map(lambda x: 1 if x >= 0.5 else 0, y_pred))
```
